```python
# started at 9:17pm
import pandas as pd
```

```python
cols = [1,2,3,4,5,6,7,8]
df = pd.read_csv("day_two_data.csv", header=None, sep=" ", names = cols)
```

```python
count = 0
for i in range(len(df)):
    posneg = 0 # 1 if increasing, -1 if decreasing
    for j in range(2,8):
        val = df.loc[i,j]
        bval = df.loc[i,j-1]
        if val != "NaN":
            if (abs(val-bval) > 0 and abs(val-bval) < 4):
                if val-bval > 0 and posneg >= 0:
                    posneg = 1
                else:
                    count += 1
                if val-bval < 0 and posneg <= 0:
                    posneg = -1
                else:
                    count += -1
            else:
                count += 1
print(count)
    
```

```python
df.fillna(0, inplace=True)
```

```python
count = 0
for i in range(len(df)):
    inner_count = 0 
    inner_tot = 0
    for j in range(2,8):
        val = df.loc[i,j]
        bval = df.loc[i,j-1]
        if val != 0:
            inner_tot += 1
            
            if ((abs(val-bval) > 0) and (abs(val-bval) < 4)):
                if val > bval:
                    inner_count += 1
                if val < bval:
                    inner_count -= 1
            else:
                count += 1
                # print(f"break at loc[{i},{j}], val = {val}, bval = {bval}")
                break
                
            # print(f"at loc[{i},{j}]")
            # print(abs(val-bval))
            # print(f"count: {count}")
        if abs(inner_count) < inner_tot:
            print(f"inner_count = {inner_count}, inner_tot = {inner_tot}")
            count += 1
    
print(count)
    
```
